```python
import pandas as pd
```

```python
## Load the data
missing_value = pd.read_csv("Week 11_Graded Mini Project_Dataset_houseprice.csv")

```

```python
## Identify and handle missing values appropriately.
## Check missing value

missing_value.isnull().sum()
```

```python
## Get the summary of the data
missing_value.describe().T
```

```python
missing_value.dropna(inplace=True)
```

```python
missing_value.fillna(missing_value.mean(), inplace=True)
```

```python
# missing_value.fillna(missing_value.mode().iloc[0], inplace=True)

categorical_columns = missing_value.select_dtypes(include=['object']).columns  # Select categorical columns

for column in categorical_columns:
    mode_value = missing_value[column].mode()
    if not mode_value.empty:
        missing_value[column].fillna(mode_value.iloc[0], inplace=True)
```

```python
missing_value.describe()
```

```python
# missing_value.isnull().sum()
# missing_value.head()
# missing_value.dtypes
# missing_value = missing_value.apply(pd.to_numeric, errors='coerce')
categorical_cols = missing_value.select_dtypes(include=['object']).columns
missing_value[categorical_cols] = missing_value[categorical_cols].astype('category')
```

```python
# missing_value.dtypes
## create binary columns
encoded_df = pd.get_dummies(missing_value, columns=['MSZoning', 'SaleType', 'SaleCondition'], drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
missing_value['SaleCondition'] = le.fit_transform(missing_value['SaleCondition'])

missing_value.head()
```
